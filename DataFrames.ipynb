### Missing Data

```python
dframe = DataFrame([[1,2,3],[np.nan,5,6],[7,np.nan,9],[np.nan,np.nan,np.nan]])
dframe
```

```python
new_dframe = dframe.dropna()
new_dframe
```

```python
#We can also specify to only drop rows that are complete missing all data
dframe.dropna(how='all')
```

```python
#Or we can specify to drop columns with missing data
dframe = DataFrame([[1,2,3,4],[np.nan,5,6,11],[7,np.nan,9,10],[np.nan,np.nan,np.nan,33]])
dframe
```

```python
dframe.dropna(axis=1)
```

```python
#Dropping any rows that dont have at least 2 data points
dframe.dropna(thresh=2)
```

```python
#We can also fill any NAN
dframe.fillna(111)
```

```python
#Can also fill in diff values for diff columns
dframe.fillna({0:10,1:100,2:200,3:300})
```

```python
#Can also fill in diff values for diff columns
dframe.fillna({1:300})
```

```python
#If we want to modify the exsisting object, use inplace
dframe.fillna(0,inplace=True)
dframe
```

### Merge data

```python
dmerge1 = DataFrame({'key':['X-Mas','Zebra','Yellow','Zebra','X-Mas','X-Mas'],'col_1': np.arange(6),'key2':['a','b','b','c','a','c']})
dmerge1
```

```python
dmerge2 = DataFrame(
    {'key':['Queen','Yellow','Zebra'],'col_2':[1,2,3],'col_3':['a','c','c']})
dmerge2
```

```python
pd.merge(dmerge1,dmerge2, on='key')
```

```python
pd.merge(dmerge1,dmerge2, on='key', how='left')
```

```python
pd.merge(dmerge2,dmerge1, on='key', how='left')
```

```python
pd.merge(dmerge1,dmerge2, on='key', how='right')
```

```python
pd.merge(dmerge1,dmerge2, on='key', how='outer', sort=True)
```

### Handling Duplicates

```python
ddup = DataFrame({'col1':['A','A','B','B','B'], 'col2':[2,2,2,3,3]})
ddup        
```

```python
#Use duplicated to find duplicates
ddup.duplicated()
```

```python
ddup.drop_duplicates()
```

```python
ddup.drop_duplicates(keep='last')
```

```python
## You can filter which duplicates to drop by a single column
## ddup.drop_duplicates(['col1'])
ddup.drop_duplicates(subset=['col1'])
```

### Mapping

```python
## Mapping
dmap = DataFrame({'state':['Kerala','Tamilnadu','Karnataka','Telangana'],'population':[1000,2500,2700,1300]})
dmap
```

```python
capitals ={'Tamilnadu':'Chennai','Karnataka':'Bangalore','Kerala':'Trivandrum','Telangana':'Hyderabad'}
capitals
```

```python
dmap['capital'] = dmap['state'].map(capitals)
dmap
```

## Group By

```python
dgrp = pd.read_csv('nba.csv')
dgrp
```

```python
grp = dgrp.groupby('Team')
```

```python
for team,group in grp:
    print(team,group[['Name','Position','Age']])
```

```python
grp.get_group('Atlanta Hawks')
```

```python
grp2 = dgrp.groupby(['Team','Position'])
```

```python
for team, group in grp2:
    print(team, group)
```

```python
grp2.get_group(('Atlanta Hawks','PG'))
```
