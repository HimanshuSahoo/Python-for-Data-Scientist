```python
## Naive Bayes Classifier Algorithm : It is a supervised algorithm based on Bayes Theorem
## It is used to solve classification problem and main text classification that includes a high dimensional dataset
## It is a probabilistic classifier which mean it predicts on the basis of the probability of an object
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(100,2, centers=2, random_state =2, cluster_std = 1.5)
```

```python
plt.scatter(X[:,0], X[:,1],c=y, s=50, cmap = 'summer');
```

```python
from sklearn.naive_bayes import GaussianNB
model_GNB = GaussianNB()
model_GNB.fit(X,y)
```

```python
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18]*rng.rand(2000,2)
ynew = model_GNB.predict(Xnew)
```

```python
plt.scatter(X[:,0], X[:,1],c=y, s=50, cmap = 'summer')
lim = plt.axis()
plt.scatter(X[:,0], X[:,1],c=y, s=50, cmap = 'summer' , alpha = 0.1)
plt.axis(lim);
```

```python
yprob = model_GNB.predict_proba(Xnew)
yprob[-10:].round(3)
```
